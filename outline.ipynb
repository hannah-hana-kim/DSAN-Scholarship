## Load the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/final_data.csv')
df
```

```python
vwc = pd.read_csv('./data/vwc_data.csv')
vwc
```

```python
# drop long and lat columns
vwc = vwc.drop(['long', 'lat'], axis=1)
vwc
```

## Time Series data

```python
# drop the NaN values
vwc = vwc.dropna()
# reset index
vwc = vwc.reset_index(drop=True)
vwc
```

```python
# group the data by year and rcp column and get the average of the other columns
vwc2 = vwc.groupby(['year', 'rcp'], as_index=False).mean()
vwc2
```

```python
# plot the data where the x-axis is the year and the y-axis is the average of the other columns and the label is rcp
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=vwc2, x='year', y='vwc_winter_whole', hue='rcp')
sns.lineplot(data=vwc2, x='year', y='vwc_summer_whole', hue='rcp')
plt.title('Volumetric Water Content')
plt.show()
```

## Heatmap

```python
#  create a heatmap
plt.figure(figsize=(10, 8))
# copy the data
vwc_copy = vwc.copy()
# convert the rcp column to numerical values
vwc_copy['rcp'] = vwc_copy['rcp'].map({'historical': 0, '4.5': 1, '8.5': 2})

# numerical_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(vwc_copy.corr(), annot=True, cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
```
